## N63. Temperature Conversion Table
>Write a program that displays a temperature conversion table for degrees Celsius and
degrees Fahrenheit. The table should include rows for all temperatures between 0
and 100 degrees Celsius that are multiples of 10 degrees Celsius. Include appropriate
headings on your columns. The formula for converting between degrees Celsius and
degrees Fahrenheit can be found on the internet

```python
def convert(temp, unit):
    unit = unit.lower()
    if unit == "c":
        temp = 9.0 / 5.0 * temp + 32
        return "%s degrees Fahrenheit"% temp
    if unit == "f":
        temp = (temp - 32)  / 9.0 * 5.0
        return "%s degrees Celsius"% temp
 

intemp = int(input("What is the temperature?\n"))
inunit = str(input("Please enter the unit of measure (f or c):\n"))
 
print(convert(intemp, inunit))
```

## N67. Admission Price


- 2  : 0 dollor
- 3  12 : 14 dollors
- 65  :18 dollors
- others : 23 dollors

```python
def admission(age):
    if age <= 2:
        return 0
    elif 3 <= age <= 12:
        return 14
    elif 65 <= age:
        return 18
    else:
        return 23

people = []
while True:
    age = input()
    if age is '':
        break
    try:
        age = int(age)
    except:
        continue
    
    people.append(age)

price_list = []
for i in people:
    price = admission(i)
    price_list.append(price)
    
# print(price_list)

total = 0
for i in price_list:
    total += i

print('Total Price is :', round(total/len(price_list),2))
    
```

## N72. Is a String a Palindrome?
> A string is a palindrome if it is identical forward and backward. For example anna,
civic, level and hannah are all examples of palindromicwords. Write a program
that reads a string from the user and uses a loop to determines whether or not it is a
palindrome. Display the result, including a meaningful output message.

```python
# while True:
#     s = input()
#     if s is '':
#         break
#     try:
#         s = list(s)
#     except:
#         continue
s = input()

def isPalindrome(s):
    reverse_list = []
    for i in range(len(s)-1, -1, -1):
        reverse_list.append(s[i])
    return reverse_list

result = isPalindrome(s)
# print(result)
if list(s) == result:
    print('>> It is Palindrome!!')
else:
    print('>> Nope!!')
```
