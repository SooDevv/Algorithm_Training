```python
import random
```

```python
# roll() 의 return 값이 total 

die_dict = {}
for i in range(2, 13):
    die_dict[i] = 0

min = 1
max = 6

def roll():
    i = 0
    while i < 5:
        first = random.randrange(min, max + 1)
        second = random.randrange(min, max + 1)
        
        die_sum = first + second
        
        die_dict[die_sum] += 1
        i += 1
    return die_dict




roll()
```
